```python
import pandas as pd
import numpy as np

# Question 1: Creating and displaying a one-dimensional array-like object using Pandas
data = pd.Series([1, 3, 5, 7, 9])
print("1. One-dimensional array-like object:")
print(data)
print()

# Question 2: Converting a Pandas Series to a Python list and displaying its type
data_list = data.tolist()
print("2. Converted Python list:", data_list)
print("   Type of converted list:", type(data_list))
print()

# Question 3: Performing arithmetic operations on two Pandas Series
series1 = pd.Series([2, 4, 6, 8, 10])
series2 = pd.Series([1, 3, 5, 7, 9])

addition = series1 + series2
subtraction = series1 - series2
multiplication = series1 * series2
division = series1 / series2

print("3. Arithmetic Operations:")
print("   Addition:")
print(addition)
print("   Subtraction:")
print(subtraction)
print("   Multiplication:")
print(multiplication)
print("   Division:")
print(division)
print()

# Question 4: Comparing elements of two Pandas Series
comparison = series1 == series2
print("4. Comparison result:")
print(comparison)
print()

# Question 5: Converting a dictionary to a Pandas Series
original_dict = {'a': 100, 'b': 200, 'c': 300, 'd': 400, 'e': 800}
converted_series = pd.Series(original_dict)
print("5. Converted series:")
print(converted_series)
print()

# Question 6: Converting a NumPy array to a Pandas Series
numpy_array = np.array([10, 20, 30, 40, 50])
converted_series = pd.Series(numpy_array)
print("6. Converted Pandas series:")
print(converted_series)
```

```python

# Sample Python dictionary data and list labels
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# 7. Select the rows where the number of attempts in the examination is greater than 2
df = pd.DataFrame(exam_data, index=labels)
print("7. Number of attempts in the examination is greater than 2:")
print(df[df['attempts'] > 2])
print()

# 8. Count the number of rows and columns of a DataFrame
print("8. Number of Rows:", len(df))
print("   Number of Columns:", len(df.columns))
print()

# 9. Select the rows where the score is missing, i.e., is NaN
print("9. Rows where score is missing:")
print(df[df['score'].isnull()])
print()

# 10. Select the rows where the score is between 15 and 20 (inclusive)
print("10. Rows where the score is between 15 and 20 (inclusive):")
print(df[(df['score'] >= 15) & (df['score'] <= 20)])
```
